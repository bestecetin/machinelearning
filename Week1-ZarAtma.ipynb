```python
import random
def bir():
    print("[keçi] [2] [3]")
def iki():
    print("[1] [keçi] [3]")
def uc():
    print("[1] [2] [keçi]")

win=0
ort=0
for y in range(11):
    for x in range(1001):
        keci1 = random.randint(1, 3)
        keci2 = random.randint(1, 3)
        while keci1== keci2:
            keci2=random.randint(1, 3)
        araba=random.randint(1, 3)
        while araba==keci2 or araba==keci1:
            araba=random.randint(1, 3)
        #print(keci1,keci2,araba)
        #print([1],[2],[3])
        secim=random.randint(1, 3)#int(input("Bir kapı seçiniz:"))
        if secim==araba and secim==1:
           rnd=random.randint(2, 3)
           if rnd==2:
               iki()
           else:
               uc()
           win+=1
        if  secim==araba and secim==2:
            rnd=random.randint(2, 3)
            if rnd==2:
                bir()
            else:
                uc()
            win += 1
        if  secim==araba and secim==3:
            rnd=random.randint(2, 3)
            if rnd==2:
                bir()
            else:
                iki()
            win += 1
        if secim!=araba and secim==1:
            if araba==2:
                uc()
            else:
                iki()
        if secim!=araba and secim==2:
            if araba==1:
                uc()
            else:
                bir()
        if secim!=araba and secim==3:
            if araba==1:
                iki()
            else:
                bir()
    print(win)
    ort+=win
    win=0
print(ort/10)
```
